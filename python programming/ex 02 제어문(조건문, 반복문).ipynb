#### 연습문제 1
- 두 수를 입력 받아 크기를 비교하는 프로그램을 만들어보자
- int(문자열) : 문자열을 정수로 변환
- float(문자열) : 문자열을 실수로 변환

```python
num1 = int(input('첫 번쨰 수 : '))
num2 = int(input('두 번쨰 수 : '))

if num1 > num2 :
    print(num1,'가 큽니다.')
elif num2 > num1 :
    print(num2,'가 큽니다.')
else :
    print('두 수의 크기는 같습니다.')

print(num1, num2)
```

### 연습문제 2


```python
score = int(input('점수를 입력해주세요.'))
if score >= 90 :
    print('%d점은 A학점 입니다.' % score)
elif score >= 80 :
    print(f'{score}점은 B학점 입니다.')
elif score >= 70 :
    print(f'{score}점은 C학점 입니다.')
elif score >= 60 :
    print('{}점은 D학점 입니다.'.format(score))
else :
    print(f'{score}점은 F학점 입니다.')
```

## 연습문제 3

```python
score_list = [90,45,70,60,55]
for i in range(5) :
    if score_list[i] >= 60 :
        print(f'{i+1}번 학생은 합격입니다.')
    else : 
        print(f'{i+1}번 학생은 불합격입니다,')

```

```python
# 선생님 풀이
score_list = [90,45,70,60,55]
cnt = 0
for s in score_list :
    cnt+=1
    rs = "합격" if s >= 60 else "불합격"
    print(f'{cnt}학생은 {rs}입니다.')
```

```python
score_list = [90,45,70,60,55]
for i in range(len(score_list)) :
    rs = "합격" if score_list[i] >= 60 else "불합격"
    print(f"{i+1}번 학생은 {rs}입니다.")
```

```python
score_list = [90,45,70,60,55]
for i,s in enumerate(score_list) :
    rs = "합격" if s >= 60 else "불합격"
    print(f"{i+1}번 학생은 {rs}입니다.")
```

### 연습문제 4 

```python
for i in range (5) :
    print(i, end=" ")
```

```python
num = int(input('정수 입력 >> '))
arr = []
for i in range(num+1) :
    if num % (i+1) == 0 :
        arr.append(i+1)
print(arr)
```

```python
# 선생님 풀이
num = int(input("정수 입력 >> "))
print(f"{num}의 약수 :",end=" ")
for i in range(1,num+1):
    if num % i == 0 :
        print(i,end=" ")
```
