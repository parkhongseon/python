## 변수의 scope

### 전역변수와 지역변수 구분

```python
num = 10 # 전역변수
def abc() :
    # 대입을 하게되면 지역변수라고 인식하게됨. 읽기만 하면 아무 상관 없음
    num = 0 # 지역변수
    num +=1
    print(num)
abc()
```

### 함수 내에서 전역변수 접근

```python
num = 10 # 전역변수
def abc2() :
    global num # 지역 변수 num이 아닌 전역변수 num을 가리킴.
    num +=1
    print(num)
abc2()
```

## 가변 매개변수

```python
def add(*args):
    print(args)
    for item in args :
        print(item, end=" ")
    
    
    
    
add(1,2,3) #튜플로 받아들이게 됨.
add(5,6,7,8,9,10)
```

```python
def abc(**args):
    print(args)


abc(a=1,b=2,c=3)
abc(name="hhd", address="광주광역시")
```
