### 리스트 연산

```python
num = [1,2,3,4]
num2 = [5,6,7,8]

#덧셈 연산
print(num+num2)

#곱셈 연산
# print(num*3)
```

#### 리스트 값 추가

```python
num.append(8)
num.append(9)
print(num)
```

####  리스트 중간에 값 추가

```python
num.insert(1,5)
print(num)
```

```python
num[2] = 10
print(num)
```

```python
num[0:2] = [7]
print(num)
```

#### 리스트 값 삭제 

```python
del num[3]
print(num)
```

```python
del num[2:4]
print(num)
```

```python
# 값을 중심으로 삭제
num.remove(10)
print(num)
```

```python
data = num.pop()
print(data)
print(num)
```
