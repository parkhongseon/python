### matplotlib 사용하기

```python
import matplotlib.pyplot as plt
#한글폰트 설정
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="C:/Windows/fonts/malgun.ttf").get_name()
rc('font',family=font_name)
```

####  line 그래프 그리기

```python
# 1. 그림판 생성
plt.figure(figsize=(10,5)) # 가로세로비율
# 2. 원하는 그래프 그리기
x = [1,2,3]
y = [4,5,6]
plt.plot(x,y,ls=':', marker="*", label="초당옥수수")

x2 = [1,2,3]
y2 = [1,2,3]
plt.plot(x2, y2, color='red', marker="^", markersize=15, label="찰옥수수")

plt.grid()
plt.xlabel('수확시기')
plt.ylabel('수확량')
plt.title("옥수수 생산량")
plt.legend()
plt.xticks([1, 1.25, 1.5, 1.75, 2, 2.25, 2.5, 2.75, 3])
plt.ylim(0,4)

# 3. 그림을 다 그렸다는 신호
plt.show()
```
