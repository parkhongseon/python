#### 조건에 맞는 데이터 접근하기
- 불리언 인덱싱 : 논리 연산자(T/F)를 활용하여 T/F인 결과를 출력하고 
  True에 해당하는 데이터만 접근
- 특정 기준에 의해 접근하는 방법


```python
# random.choice : 랜덤으로 요소 하나를 취득하겠다
# random.sample : 랜덤으로 요소 여러개를 취득, 중복 없음
# random.choices : 랜덤으로 요소 여러개를 취득하지만 중복이 있음.
```

```python
# 불리언 배열 생성
sample_arr = [True, False]

np.random.seed(2)
bool_arr = np.random.choice(sample_arr, size = 10) # True, False : 10개 요소
display(bool_arr)

tmp_data = np.arange(10) # 0 ~ 9 까지 10개 요소
display(tmp_data)

tmp_data[bool_arr]
```

```python
# 1. 성적을 담고 있는 배열
# 2. 학생들의 번호를 담고 있는 배열

np.random.seed(2)
score = np.random.randint(50, 100, size = 31)
std_num = np.arange(1, 32)
display(score)
display(std_num)
```

```python
# 논리 연산자 사용 , 80점 이상인 성적 확인

score >= 80
score [score >= 80] 
```

```python
# 70점 미만인 데이터는 몇 개일까
score[score < 70].size
```

```python
# 성적이 60점 이상인 학생의 번호 출력
std_num[score >= 60]
std_num[score < 60] # 60점 미만인 학생들의 번호
```

```python
# 데이터 차원
display(data.ndim)
# 데이터 크기 [행, 열]
display(data.shape)
# 전체 데이터의 요소 갯수
display(data.size)
```

####  전체 평점 평균 구하기

```python
all_mean = np.mean(data[:,2])
all_mean
```

```python
# 1. 사용자 아이디만 접근 
user_id = data[: , 0]
user_id
```

```python
# 2. 사용자 아이디 중복 없이 확인
u_id = np.unique(user_id)
display(u_id)
display(np.unique(user_id, return_counts=True))
```

```python
# 3. 사용자 아이디가 1인 데이터의 평점 평균 구해보기
    # data[np.where(data[:,0] == 1)] 
u_id1 = data[data[:, 0] == 1]
u_id1[:,2].mean()
```

```python
# 4. 모든 사용자의 평점에 접근하여 평점 평균 구해보기
# 저장형식 [사용자 아이디, 평점 평균]
data[data[:, 0] == 3][:, 2].mean()
# hint 1 : 사용자별 평점 평균을 저장한 빈 리스트 선언
# hint 2 : for 반복문 : 사용자 아이디 전체를 담고 있는 변수 확인
# hint 3 : 각 사용자별 평균값
# hint 4 : 리스트에 [사용자 아이디, 평균 값] 형식으로 추가

result= []
for i in u_id :
    rating_mean = data[data[:, 0] == i][:, 2].mean()
    result.append([i, rating_mean])
```

```python
# 5. 평점 평균이 4점 이상인 사용자는 몇 명인지 확인 (불리언 인덱싱)
# hint 1.불리언 인덱싱을 적용하기 위해 array() 변환
# hint 2. array 에 접근해서 평점의 평균이 4점 이상인 데이터 확인
# hint 3. 평점의 평균이 4점 이상인 사용자 수를 확인 .size
user_mean = np.array(result, dtype= np.float64)
user_mean
```

```python
user_mean[user_mean[:, 1] >= 4][:, 0].size
```

```python
user_mean[np.where(user_mean[:,1]>=4)][:, 0].size # where 는 인덱스 번호 반환
```
